```python
# Escriba la solución aquí
Nombre = input("Digite su nombre: ").title().strip()
Monto = float(input("Digite el monto de la compra: "))
Cliente = input("Digite si usted es frecuente o no frecuente: ").title().strip()

if Cliente=="Frecuente" and Monto > 50000:
    descuento = Monto * 0.10
elif Cliente=="Frecuente" or Monto > 100000:
    descuento = Monto * 0.05
else:
    descuento=0
totalfinal = Monto - descuento

print(f"{Nombre}, usted es cliente {Cliente} y su descuento es {descuento}% y el total a pagar es de ${totalfinal}")
```
