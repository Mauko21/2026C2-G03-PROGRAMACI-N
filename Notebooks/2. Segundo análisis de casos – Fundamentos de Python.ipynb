# Problema #1: Calculadora de propina inteligente

## Problema

| 1 | Calculadora de propina inteligente | Solicita total de cuenta y porcentaje de propina. Calcula propina y total final. | `float`, operaciones, f-strings |

## Entrada

- Total de la cuenta
Porcentaje que usted quiere dar

## Proceso

- Multiplicar el total de la cuenta entre el resultado del porcentaje de propina dividido entre 100
despues, para calcular el total final solo sumar la propina con el total de la cuenta

## Salida esperada

- El total final de a pagar
Y si el numero lo digita en letras que le salga error.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | | | |
| Caso límite | | | |
| Caso inválido | | | |


```python
# Escriba la solución aquí
try:
    TotalCuenta = float(input("Digite el total pagado de la cuenta: "))
    PorcentajePropina = float(input("Digite el porcentaje de propina que quiere dar: "))

    Propina = TotalCuenta * (PorcentajePropina/100)
    TotalFinal = TotalCuenta + Propina
    print(f"El total que se debe de pagar es {TotalFinal}, y la propina que debe de dar es de {Propina}")
except ValueError:
    print("Digite un número entero, no en letras.")
```

# Problema #2: Detector de número par/impar

## Problema

| 2 | Detector de número par/impar | Solicita un entero e indica si es par o impar. | `int`, `%`, `if/else` |

## Entrada

- digitar un numero cualquiera que no sea negativo

## Proceso

- dividir numero entre 2, si el numero da numero entero, es par, en cambio si da con decimales es impar.

## Salida esperada

- Que si el numero es par, que salga en la ventana de comandos que es par.
Si es negativo, que salga en la ventana de comandos que es negativo y que tiene que digitar otro.
Si es impar, que salga en la ventana de comandos que es impar.
Si el numero esta escrito en letras, que salga error.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | | | |
| Caso límite | | | |
| Caso inválido | | | |


```python
# Escriba la solución aquí
try:
    num = int(input("Digite un número: "))
    if num%2==0:
        print("Su numero es par")
    elif num<0:
        print("Su numero es negativo, digite otro.")
    else:
        print("Su numero es impar")
except ValueError:
    print("Digite un número entero, no en letras.")
```

# Problema #3: Clasificador de edad

## Problema

| 3 | Clasificador de edad | Solicita edad y clasifica como menor, adulto o adulto mayor. | `int`, `if/elif/else` |

## Entrada

- Digitar la edad

## Proceso

- A partir de un if...
Si la edad es negativa, no se puede entonces, tiene que digitar que es negativa.
Si es menor de 18, que salga que es menor de edad.
Si es mayor o igual a 18 pero que no supere los 65 años, que salga que es mayor de edad.
Si es mayor o igual a 65 años, que salga que usted es un adulto mayor.

## Salida esperada

- La edad y si usted es mayor de edad, menor de edad o adulto mayor.

Y si el numero esta escrito en letras, que le salga error.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | | | |
| Caso límite | | | |
| Caso inválido | | | |


```python
# Escriba la solución aquí
try:
    Edad = int(input("Digite su edad actual:"))
    if Edad<0:
        print("La edad no puede ser negativa")
    elif Edad<18:
        print(f"Su edad es {Edad} por lo que usted es menor de edad")
    elif Edad>=18:
        print(f"Su edad es {Edad} por lo que usted es mayor de edad")
    elif Edad>=65:
        print(f"Su edad es {Edad} por lo que usted es un adulto mayor")
    
    
except ValueError:
    print("Digite un numero entero, no en letras.")
```

# Problema #4: Validador de compra con descuento

## Problema

Escriba aquí la descripción del problema.
| 4 | Validador de compra con descuento | Solicita cliente, monto y si es cliente frecuente. Aplica descuento según reglas. | strings, `float`, `and/or`, f-strings |


## Entrada

- Nombre, monto a pagar y el tipo de cliente que sea

## Proceso

- si dice que es cliente frecuente y el monto a pagar es de más de 50000, que se haga descuento de un 10%
Si es cliente frecuente pero no supera los 50000 o si su monto a pagar es de 100000, que se haga descuento de 5%
Pero si usted no supera los 50000 y no es cliente frecuente, no se aplica descuento

## Salida esperada

- El nombre, tipo de cliente, el descuento que se le hace, y el total a pagar con el descuento aplicado.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | | | |
| Caso límite | | | |
| Caso inválido | | | |


```python
# Escriba la solución aquí
Nombre = input("Digite su nombre: ").title().strip()
Monto = float(input("Digite el monto de la compra: "))
Cliente = input("Digite si usted es frecuente o no frecuente: ").title().strip()

if Cliente=="Frecuente" and Monto > 50000:
    descuento = Monto * 0.10
elif Cliente=="Frecuente" or Monto > 100000:
    descuento = Monto * 0.05
else:
    descuento=0
totalfinal = Monto - descuento

print(f"{Nombre}, usted es cliente {Cliente} y su descuento es {descuento}% y el total a pagar es de ${totalfinal}")
```
